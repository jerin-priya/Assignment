```python
import matplotlib.pyplot as plt
```

```python
def bresenham_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    # If slope is greater than 1, swap the roles of x and y
    if abs(dy) > abs(dx):
        x1, y1 = y1, x1
        x2, y2 = y2, x2
        dx = x2 - x1
        dy = y2 - y1
        swapped = True
    else:
        swapped = False
        
    # Initialize variables
    p = 2 * abs(dy) - abs(dx)
    two_dy = 2 * abs(dy)
    two_dy_minus_dx = 2 * (abs(dy) - abs(dx))
    x = x1
    y = y1
    points = [(x, y)]

    # Determine the direction of increment along y
    if dy < 0:
        y_inc = -1
    else:
        y_inc = 1

    # Iterate through the x values
    for i in range(abs(dx)):
        x += 1
        if p < 0:
            p += two_dy
        else:
            y += y_inc
            p += two_dy_minus_dx
        if swapped:
            points.append((y, x))
        else:
            points.append((x, y))
    return points
```

```python
cases = [((1, 1), (8, 4)), ((1, 1), (4, 8))]
```

```python
fig, axes = plt.subplots(1, len(cases), figsize=(10, 5))

for ax, ((x1, y1), (x2, y2)) in zip(axes, cases):
    ax.set_title(f"Line ({x1},{y1}) to ({x2},{y2})")
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.grid(True)
    
    points = bresenham_line(x1, y1, x2, y2)
    x_values = [point[0] for point in points]
    y_values = [point[1] for point in points]
    
    ax.plot(x_values, y_values, marker='o')
    ax.plot([x1, x2], [y1, y2], linestyle='--', color='gray')  # Plot the actual line
    
plt.tight_layout()
plt.show()
```
